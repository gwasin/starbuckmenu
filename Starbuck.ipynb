```python
# 17.K-Means Clustering organize data as Healthy, Balanced, High Sugar
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
features = df[['Calories', 'Sugar (g)']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
intertia = []
for k in range(1, 6):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_features)
    intertia.append(kmeans.inertia_)
plt.plot(range(1, 6), intertia)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
kmeans_model = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster'] = kmeans_model.fit_predict(scaled_features)
```

```python
# 20.PCA (Principal Component Analysis)
from sklearn.decomposition import PCA
df['Calories'] = pd.to_numeric(df['Calories'], errors='coerce')
df['Total fat (g)'] = pd.to_numeric(df['Total fat (g)'], errors='coerce')
df['Sugar (g)'] = pd.to_numeric(df['Sugar (g)'], errors='coerce')
df['Carbohydrates (g)'] = pd.to_numeric(df['Carbohydrates (g)'], errors='coerce')
df = df.dropna(subset=['Calories', 'Total fat (g)', 'Sugar (g)', 'Carbohydrates (g)'])
features = ['Calories', 'Total fat (g)', 'Sugar (g)', 'Carbohydrates (g)']
x = df.loc[:, features].values
x = StandardScaler().fit_transform(x)
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x)

principal_df = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])
print("Principal Components:\n", principal_df)
print("Explained Variance Ratio:\n", pca.explained_variance_ratio_)

loadings = pd.DataFrame(pca.components_, columns=features, index=['PC1', 'PC2'])
print("Component Loading:\n", loadings)
```

```python
# 21.RandomForest
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
x = df[['Total fat (g)', 'Sugar (g)']]
y = df['Calories']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
rf_predictions = rf_model.predict(x_test)
rf_r2 = rf_model.score(x_test, y_test)
print(f"RandomForest Regressor R-squared: {rf_r2:.2f}")
```

```python
# 22.Classifier (Logistic / XGBoost)
df['Label'] = ((df['Calories'] < 250) & (df['Sugar (g)'] < 15)).astype(int)
x = df[['Calories', 'Sugar (g)']]
y = df['Label']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
log_reg_model = LogisticRegression()
log_reg_model.fit(x_train, y_train)
y_pred_log_reg = log_reg_model.predict(x_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_log_reg))
print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_log_reg))
```

```python
# XGBoost
import xgboost as xgb
xgb_model = xgb.XGBClassifier(objective='binary:logistic', use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(x_train, y_train)
y_pred_xgb = xgb_model.predict(x_test)
print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("XGBoost Classification Report:\n", classification_report(y_test, y_pred_xgb))
```

```python
# 23.Recommendation System with Content-Based Filtering
from sklearn.metrics.pairwise import euclidean_distances
def recommend_low_cal_sugar_menus(menu_name, df, num_recommendations=5):
    if menu_name not in df['Product Name'].values:
        print(f"Not Found Menu'{menu_name} in Data")
        return df
    target_menu_features = df[df['Product Name'] == menu_name][['Calories', 'Sugar (g)']]
    other_menus_df = df[df['Product Name'] != menu_name].copy()
    other_menu_features = other_menus_df[['Calories', 'Sugar (g)']].values
    distances = euclidean_distances(target_menu_features, other_menu_features)
    other_menus_df['distance'] = distances[0]
    recommended_menues = other_menus_df.sort_values(by='distance', ascending=True)
    return recommended_menues.head(num_recommendations)
recommended_list = recommend_low_cal_sugar_menus('Brewed Coffee', df)
print(recommended_list[['Product Name', 'Calories', 'Sugar (g)']])
```

```python
# 24.Use Anomaly Detection Which Menues are abnormal
from sklearn.ensemble import IsolationForest
x = df[['Calories', 'Label Wt (g)']]
model = IsolationForest(contamination=0.1, random_state=42)
model.fit(x)
df['anomaly'] = model.predict(x)
abnormal_menus = df[df['anomaly'] == -1]
abnormal_re = abnormal_menus.sort_values(by='Calories', ascending=False)
print(abnormal_re[['Product Name', 'Calories']])
```
